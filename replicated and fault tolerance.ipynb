```python
json_file_path='meta-California.json'
def upsert_document(collection, document_id, document, max_retries=3, retry_delay=1):
    for attempt in range(max_retries):
        try:
            # Perform the upsert operation on the cluster
            result = collection.upsert(document_id, document)

            # Check the result
            if result.success:
                break
            else:
                print(f'Retry {attempt + 1}: Error uploading document - {result.error}')
                time.sleep(retry_delay)

        except AmbiguousTimeoutException as e:
            print(f'Retry {attempt + 1}: AmbiguousTimeoutException - {e}')
            time.sleep(retry_delay)

        except Exception as e:
            print(f'Retry {attempt + 1}: Error - {e}')
            time.sleep(retry_delay)

# Read from the JSON file and attempt to upsert each document

with open(json_file_path, 'r') as file:
    for line in file:
        try:
            document = json.loads(line)
            document_id = document.get('gmap_id')  # Use a unique identifier as the document ID

            upsert_document(collection, document_id, document)

        except json.JSONDecodeError as e:
            print(f"Error decoding JSON: {e}")
```

```python
query2 = f"SELECT * FROM cal LIMIT 500"
result2 = cluster.query(query2)
data = []

for row2 in result2:
    data.append(row2)
```

```python
import folium
california_lat, california_lon = 36.7783, -119.4179

# Create a Folium map centered on California
map_center = [california_lat, california_lon]
mymap = folium.Map(location=map_center, zoom_start=6)

# Plot fabric store locations on the map
for store_data in data:
    store = store_data['cal']
    lat, lon = store['latitude'], store['longitude']
    folium.Marker([lat, lon], popup=store.get('name', 'Unknown'), icon=folium.Icon(color='red')).add_to(mymap)

# Display the map in the Jupyter Notebook
display(mymap)
```
